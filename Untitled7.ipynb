```python
from tkinter import *
from tkinter.scrolledtext import ScrolledText

def load():
    with open(filename.get()) as f:
        contents.delete('1.0', END)
        contents.insert(INSERT, f.read())

def save():
    with open(filename.get(), 'w') as f:
        f.write(contents.get('1.0', END))
        
top = Tk()
top.title("Simple text editor")
contents = ScrolledText()
contents.pack(side=BOTTOM, expand=True, fill=BOTH)

filename = Entry()
filename.pack(side=LEFT, expand=True, fill=X)
button_open = Button(text='Open')
button_open.pack(side=LEFT)
button_open['command'] = load

button_save = Button(text='Save')
button_save.pack(side=LEFT)
button_save['command'] = save

mainloop()
```

```python
from tkinter import *
top = Tk()
top.title('Simple calculator')

BEFORE_FIRST_NUMBER = 0
INPUT_FIRST_NUMBER = 1
RESULT_DISPLAYED = 2
BEFORE_SECOND_NUMBER = 3
INPUT_SECOND_NUMBER = 4

state = BEFORE_FIRST_NUMBER

number1 = ''
number2 = ''
operation = ''

def on_number_clicked(ch):
    global state
    global number1
    global number2
    
    if state == BEFORE_FIRST_NUMBER:
        number1 = ch
        result['text'] = number1
        state = INPUT_FIRST_NUMBER
    elif state == INPUT_FIRST_NUMBER:
        number1 = number1 + ch
        result['text'] = number1
    elif state == BEFORE_SECOND_NUMBER:
        number2 = ch
        result['text'] = number2
        state = INPUT_SECOND_NUMBER
    elif state == RESULT_DISPLAYED:
        number1 = ch
        state = INPUT_FIRST_NUMBER
        result['text'] = number1
    else:
        number2 = number2 + ch
        result['text'] = number2
    
def onclick_1():
    print('1 clicked')
    on_number_clicked('1')

def onclick_2():
    print('2 clicked')
    on_number_clicked('2')
    
def onclick_3():
    print('3 clicked')
    on_number_clicked('3')

def onclick_4():
    print('4 clicked')
    on_number_clicked('4')

def onclick_5():
    print('5 clicked')
    on_number_clicked('5')

def onclick_6():
    print('6 clicked')
    on_number_clicked('6')

def onclick_7():
    print('7 clicked')
    on_number_clicked('7')

def onclick_8():
    print('8 clicked')
    on_number_clicked('8')

def onclick_9():
    print('9 clicked')
    on_number_clicked('9')

def onclick_0():
    print('0 clicked')
    on_number_clicked('0')

def onclick_dot():
    print('dot clicked')
    on_number_clicked('.')

def onclick_operator(op):
    global state
    global number1
    global number2
    global operation
    
    if state == BEFORE_FIRST_NUMBER:
        pass
    elif state == INPUT_FIRST_NUMBER:
        state = BEFORE_SECOND_NUMBER
        operation = op
    elif state == BEFORE_SECOND_NUMBER:
        operation = op
    elif state == RESULT_DISPLAYED:
        operation = op
        state = BEFORE_SECOND_NUMBER
    else:
        pass
    
def onclick_add():
    print('add clicked')
    onclick_operator('+')
    
def onclick_mul():
    print('mul clicked')
    onclick_operator('*')

def onclick_div():
    print('div clicked')
    onclick_operator('/')

def onclick_minus():
    print('minus clicked')
    onclick_operator('-')

def onclick_run():
    global state
    global number1
    global number2
    global operation

    print('run clicked')
    if operation == '+':
        number1 = str(float(number1) + float(number2))
    elif operation == '-':
        number1 = str(float(number1) - float(number2))
    elif operation == '*':
        number1  = str(float(number1) * float(number2))
    elif operation == '/':
        number1 = str(float(number1) / float(number2))
    
    result['text'] = number1
    state = RESULT_DISPLAYED
    
result = Label(text='this is the result')
result.grid(row=1, column=1, columnspan=4)

button_div = Button(text='/', command=onclick_div)
button_div.grid(row=2, column=4)

button_7 = Button(text='7', command=onclick_7)
button_7.grid(row=3, column=1)
button_8 = Button(text='8', command=onclick_8)
button_8.grid(row=3, column=2)
button_9 = Button(text='9', command=onclick_9)
button_9.grid(row=3, column=3)
button_mul = Button(text='x', command=onclick_mul)
button_mul.grid(row=3, column=4)

button_4 = Button(text='4', command=onclick_4)
button_4.grid(row=4, column=1)
button_5 = Button(text='5', command=onclick_5)
button_5.grid(row=4, column=2)
button_6 = Button(text='6', command=onclick_6)
button_6.grid(row=4, column=3)
button_minus = Button(text='-', command=onclick_minus)
button_minus.grid(row=4, column=4)

button_1 = Button(text='1', command=onclick_1)
button_1.grid(row=5, column=1)
button_2 = Button(text='2', command=onclick_2)
button_2.grid(row=5, column=2)
button_3 = Button(text='3', command=onclick_3)
button_3.grid(row=5, column=3)
button_add = Button(text='+', command=onclick_add)
button_add.grid(row=5, column=4)

button_0 = Button(text='0', command=onclick_0)
button_0.grid(row=6, column=2)
button_dot = Button(text='.', command=onclick_dot)
button_dot.grid(row=6, column=3)
button_run = Button(text='=', command=onclick_run)
button_run.grid(row=6, column=4)

mainloop()
```
